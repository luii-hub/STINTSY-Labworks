### Evaluate using Root Mean Squared Error (RMSE)

Is there a objective way of saying if the model is doing well or not?

This is called regression metrics, and we're going to try out one now called **root mean squared error** or **RMSE**. 

$$
\operatorname{RMSE}(\hat{y}) = \sqrt{\operatorname{MSE}(\hat{y})} = \sqrt{\operatorname{M}((\hat{y}-y)^2)}
$$

$\hat{y}$ stands for predicted labels, while $y$ stands for the ground truth label

Computing it can be summarized in 4 simple steps:
1. Have a test data ready with known labels. Ideally, these data have never been used in training, but we're going to suspend that rule first and try it with our dataset.
2. Get the prediction for every test data, and **subtract** this estimate with the ground truth, and **square** the difference.
3. Get the **average** of all the squared differences (#2) of every test data. 
4. Get the **square root** of the value from #3.

Open `linear_regression.py` file and complete the `compute_RMSE()` function. This function computes the Root Mean Squared Error (RMSE) given the ground truth values and the predicted values.

Implement the `compute_RMSE()` function. Inline comments should help you in completing the contents of the function.

To check our implementation, let's compute the RMSE of the two `np.ndarray` with the same values. This should yield an RMSE of 0.0.

````python
y_sample_true = np.arange(start=0, stop=20, step=1)
y_sample_pred = np.arange(start=0, stop=20, step=1)

rmse = compute_RMSE(y_sample_true, y_sample_pred)
print('{:.2f}'.format(rmse))
````

### GPT Answer

The different predicted price you obtained from the Gradient Descent (SGDRegressor) method compared to the Analytical and PolyFit methods is not necessarily because Gradient Descent is better. Instead, it's due to the nature of how each method optimizes the model:

Key Differences:

<b> Analytical Method: </b>
The analytical solution (using the Normal Equation) computes the exact weights by solving the linear regression equation directly. This gives the most accurate and precise solution because it solves the problem mathematically and deterministically.

<b> PolyFit Method: </b>
np.polyfit() uses polynomial fitting (which also returns an exact solution for linear models). For linear regression, it gives the same results as the analytical method because it directly computes the optimal weights.

<b> Gradient Descent (SGDRegressor):</b>
Stochastic Gradient Descent (SGD) is an iterative optimization method that updates weights gradually by minimizing the loss function (typically Mean Squared Error in regression).
Since SGD is iterative, it may not always converge to the exact solution, especially if the number of iterations is limited or the learning rate is too high/low. In your case, the different predicted value could be because the SGD process hasn't fully converged to the exact solution after just a few iterations.

Additionally, the learning rate (eta0) and the random starting point of weights (if not initialized carefully) can affect the final predictions.

### So, why is the predicted price different using Gradient Descent?
Gradient Descent is an approximation method: Unlike the analytical method, which provides an exact solution, SGD approximates the solution iteratively. The predicted value may not match exactly if the algorithm hasn't fully converged to the optimal solution.

Convergence: If you increase the number of iterations (max_iter), decrease the learning rate (eta0), or adjust the other hyperparameters, the SGD prediction might get closer to the values from the analytical solution.

Conclusion:
The different result from SGD is not because its better but because its an iterative method that might not have fully converged to the optimal solution within the settings you used. The analytical method gives the most accurate result for this type of linear regression problem, but SGD can be useful in cases where an analytical solution is computationally expensive or infeasible for very large datasets.

### GPT Improv:

To improve the convergence of **Stochastic Gradient Descent (SGD)** and get closer to the analytical or PolyFit method's results, we can try adjusting the following parameters:

### Key Parameters to Adjust:
1. **`max_iter`**: Increase the number of iterations so that SGD can take more steps towards convergence.
2. **`eta0` (Learning Rate)**: The learning rate controls how big each step of the gradient descent is. If its too large, the algorithm may overshoot the optimal solution; if its too small, convergence can be slow.
3. **`tol` (Tolerance)**: This specifies the stopping criterion. By lowering this value, the model can continue running for more iterations until it gets closer to the optimal solution.
